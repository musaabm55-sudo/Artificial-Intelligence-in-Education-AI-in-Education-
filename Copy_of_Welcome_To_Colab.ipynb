```
# ÿßŸÑÿÆŸÑŸäÿ© 2: ÿ™ŸÜÿ≤ŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿ£ÿØŸàÿßÿ™ ŸÇŸäÿßÿ≥Ÿäÿ© ŸÅŸä Python.
import tensorflow as tf
import numpy as np
import os
import matplotlib.pyplot as plt
import random
import tarfile # ŸÖŸÉÿ™ÿ®ÿ© Python ŸÇŸäÿßÿ≥Ÿäÿ© ŸÑŸÖŸÑŸÅÿßÿ™ .tar
import requests # ŸÖŸÉÿ™ÿ®ÿ© Python ŸÑÿ™ŸÜÿ≤ŸäŸÑ ÿßŸÑŸÖŸÑŸÅÿßÿ™ ÿπÿ®ÿ± HTTP

from IPython.display import display, HTML, clear_output

# Download and extract the data constants
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/4Z1fwRR295-1O3PMQBH6Dg/images-dataSAT.tar"
extraction_path = "."
local_file_name = "images-dataSAT.tar"

# --- ÿØÿßŸÑÿ© ŸÇŸäÿßÿ≥Ÿäÿ© ŸÑÿ™ŸÜÿ≤ŸäŸÑ Ÿàÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ (ÿ™ÿ¨ŸÜÿ® async/await) ---
def download_and_extract_data():
    clear_output(wait=True)
    display(HTML("<b>‚è≥ ÿ¨ÿßÿ±Ÿä ÿ™ŸÜÿ≤ŸäŸÑ ÿßŸÑŸÖŸÑŸÅ...</b>"))

    # 1. ÿßŸÑÿ™ŸÜÿ≤ŸäŸÑ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ requests
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status() # Ÿäÿ™ÿ≠ŸÇŸÇ ŸÖŸÜ ŸÜÿ¨ÿßÿ≠ ÿ≠ÿßŸÑÿ© HTTP

        with open(local_file_name, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)

        display(HTML("<b>‚úÖ ÿ™ŸÖ ÿ™ŸÜÿ≤ŸäŸÑ ÿßŸÑŸÖŸÑŸÅ ÿ®ŸÜÿ¨ÿßÿ≠. ÿ¨ÿßÿ±Ÿä ÿßŸÑÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨...</b>"))

        # 2. ÿßŸÑÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ tarfile
        with tarfile.open(local_file_name, 'r') as tar:
            tar.extractall(path=extraction_path)

        # 3. ÿßŸÑÿ™ŸÜÿ∏ŸäŸÅ (ÿßÿÆÿ™Ÿäÿßÿ±Ÿä)
        os.remove(local_file_name)

        display(HTML("<b>‚úÖ ÿ™ŸÖ ÿ™ŸÜÿ≤ŸäŸÑ Ÿàÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ®ŸÜÿ¨ÿßÿ≠ ŸÅŸä ÿßŸÑŸÖÿ≥ÿßÿ±: ./images_dataSAT/</b>"))

    except requests.exceptions.RequestException as e:
        display(HTML(f"<b>‚ùå ÿÆÿ∑ÿ£ ŸÅŸä ÿßŸÑÿ™ŸÜÿ≤ŸäŸÑ ÿπÿ®ÿ± ÿßŸÑÿ•ŸÜÿ™ÿ±ŸÜÿ™: {e}</b>"))
    except tarfile.TarError as e:
        display(HTML(f"<b>‚ùå ÿÆÿ∑ÿ£ ŸÅŸä ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑŸÖŸÑŸÅ ÿßŸÑŸÖÿ∂ÿ∫Ÿàÿ∑: {e}</b>"))
    except Exception as e:
        display(HTML(f"<b>‚ùå ÿ≠ÿØÿ´ ÿÆÿ∑ÿ£ ÿ∫Ÿäÿ± ŸÖÿ™ŸàŸÇÿπ: {e}</b>"))

# --- ÿ™ÿ¥ÿ∫ŸäŸÑ ÿßŸÑÿØÿßŸÑÿ© ÿßŸÑŸÇŸäÿßÿ≥Ÿäÿ© ---
# Ÿäÿ¨ÿ® ÿ£ŸÜ ÿ™ŸÉŸàŸÜ ŸÖŸÉÿ™ÿ®ÿ© requests ŸÖÿ´ÿ®ÿ™ÿ© (ŸàŸáÿ∞ÿß Ÿäÿ≠ÿØÿ´ ŸÅŸä ÿßŸÑÿÆŸÑŸäÿ© 1)
download_and_extract_data()
```

```
# ÿßŸÑÿÆŸÑŸäÿ© 3: ÿ™ŸÜŸÅŸäÿ∞ ŸÖŸÜÿ∑ŸÇ ÿßŸÑŸÖŸáÿßŸÖ (Task 1-4).
# Define Directories and Data Constants
base_dir = './images_dataSAT/'
dir_non_agri = os.path.join(base_dir, 'class_0_non_agri')
dir_agri = os.path.join(base_dir, 'class_1_agri')
IMAGE_SIZE = (64, 64)
BATCH_SIZE = 8
SEED = 1337

print("\n--- 1. Data Setup Complete ---")

# --- Task 1: Create paths and labels ---
all_image_paths = []
labels = []

# Process class 0 (Non-Agri)
if os.path.exists(dir_non_agri):
    for filename in os.listdir(dir_non_agri):
        if filename.endswith((".jpg", ".png")):
            all_image_paths.append(os.path.join(dir_non_agri, filename))
            labels.append(0)
# Process class 1 (Agri)
if os.path.exists(dir_agri):
    for filename in os.listdir(dir_agri):
        if filename.endswith((".jpg", ".png")):
            all_image_paths.append(os.path.join(dir_agri, filename))
            labels.append(1)

print(f"Total images loaded: {len(all_image_paths)}")

# --- Task 2: Create temp list, shuffle, and print 5 samples ---
if all_image_paths:
    temp = list(zip(all_image_paths, labels))
    random.shuffle(temp)

    print("\n--- 2. Task 2: 5 Random Samples (Path, Label) ---")
    for path, label in random.sample(temp, min(5, len(temp))):
        print(f"Label: {label} | Path: {path}")

# --- Define the Custom Data Generator Function ---
def custom_data_generator(image_paths, labels, batch_size, target_size=(64, 64)):
    num_samples = len(image_paths)
    while True:
        indices = np.arange(num_samples)
        np.random.shuffle(indices)
        shuffled_paths = np.array(image_paths)[indices]
        shuffled_labels = np.array(labels)[indices]

        for offset in range(0, num_samples, batch_size):
            batch_paths = shuffled_paths[offset:offset+batch_size]
            batch_labels = shuffled_labels[offset:offset+batch_size]

            batch_images = []
            for path in batch_paths:
                img = tf.keras.utils.load_img(path, target_size=target_size)
                img_array = tf.keras.utils.img_to_array(img)
                batch_images.append(img_array)

            yield np.array(batch_images) / 255.0, np.array(batch_labels)

# --- Task 3: Test and Visualize the Custom Generator ---
if all_image_paths:
    my_generator = custom_data_generator(all_image_paths, labels, batch_size=BATCH_SIZE)
    images_batch_custom, labels_batch_custom = next(my_generator)

    print("\n--- 3. Task 3: Custom Generator Test ---")
    print(f"Batch Image Shape (Custom Gen): {images_batch_custom.shape}")

    plt.figure(figsize=(10, 5))
    plt.suptitle("Custom Data Generator Batch Visualization", fontsize=14)
    for i in range(BATCH_SIZE):
        ax = plt.subplot(2, 4, i + 1)
        plt.imshow(images_batch_custom[i])
        plt.title(f"Label: {labels_batch_custom[i]}")
        plt.axis("off")
    plt.show()


# --- Task 4 & Performance Configuration: Keras Built-in Utility ---
try:
    # Create Training Dataset
    train_ds = tf.keras.utils.image_dataset_from_directory(
        base_dir,
        labels='inferred',
        label_mode='int',
        validation_split=0.2,
        subset='training',
        seed=SEED,
        image_size=IMAGE_SIZE,
        batch_size=BATCH_SIZE
    )

    # Create Validation Dataset (Task 4)
    val_ds = tf.keras.utils.image_dataset_from_directory(
        base_dir,
        labels='inferred',
        label_mode='int',
        validation_split=0.2,
        subset='validation',
        seed=SEED,
        image_size=IMAGE_SIZE,
        batch_size=BATCH_SIZE
    )

    print("\n‚úÖ ÿ™ŸÖ ÿ•ŸÜÿ¥ÿßÿ° ŸÖÿ¨ŸÖŸàÿπÿßÿ™ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ (Training & Validation).")

    # Configure for Performance
    AUTOTUNE = tf.data.AUTOTUNE
    data_augmentation = tf.keras.Sequential([
        tf.keras.layers.RandomFlip("horizontal_and_vertical"),
        tf.keras.layers.RandomRotation(0.2),
    ])

    train_ds = train_ds.map(lambda x, y: (data_augmentation(x, training=True), y))
    train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
    val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)

    print("‚úÖ ÿ™ŸÖ ÿ•ÿπÿØÿßÿØ ŸÖÿ¨ŸÖŸàÿπÿ© ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÑÿ£ÿØÿßÿ° ÿπÿßŸÑŸç.")

except Exception as e:
    print(f"‚ùå ÿÆÿ∑ÿ£ ŸÅŸä ÿ™ŸÜŸÅŸäÿ∞ Keras Utility. ÿ™ÿ£ŸÉÿØ ŸÖŸÜ ÿ•ŸÉŸÖÿßŸÑ ÿßŸÑÿÆŸÑŸäÿ© 2: {e}")
```

```
from torchvision import transforms

# üö® Ÿäÿ¨ÿ® ÿ£ŸÜ ÿ™ŸÉŸàŸÜ ÿßŸÑÿµŸàÿ±ÿ© ŸÖŸÜ ŸÜŸàÿπ PIL Image ÿ£Ÿà Tensor ŸÇÿ®ŸÑ ÿ™ÿ∑ÿ®ŸäŸÇ ÿßŸÑÿ™ÿ≠ŸàŸäŸÑÿßÿ™
custom_transform = transforms.Compose([
    # 1. ÿ™ÿ∫ŸäŸäÿ± ÿßŸÑÿ≠ÿ¨ŸÖ (Resize): ÿßŸÑÿ≠ÿ¨ŸÖ 64x64
    transforms.Resize((64, 64)),

    # 2. ÿßŸÑÿßŸÜŸÇŸÑÿßÿ® ÿßŸÑÿ£ŸÅŸÇŸä ÿßŸÑÿπÿ¥Ÿàÿßÿ¶Ÿä (Random Horizontal Flip): ÿßÿ≠ÿ™ŸÖÿßŸÑ 0.5
    transforms.RandomHorizontalFlip(p=0.5),

    # 3. ÿßŸÑÿßŸÜŸÇŸÑÿßÿ® ÿßŸÑÿπŸÖŸàÿØŸä ÿßŸÑÿπÿ¥Ÿàÿßÿ¶Ÿä (Random Vertical Flip): ÿßÿ≠ÿ™ŸÖÿßŸÑ 0.2
    transforms.RandomVerticalFlip(p=0.2),

    # 4. ÿßŸÑÿØŸàÿ±ÿßŸÜ ÿßŸÑÿπÿ¥Ÿàÿßÿ¶Ÿä (Random Rotation): ÿ≤ÿßŸàŸäÿ© 45 ÿØÿ±ÿ¨ÿ© (ŸÅŸä ŸÖÿØŸâ -45 ÿ•ŸÑŸâ 45)
    transforms.RandomRotation(45),

    # 5. ÿßŸÑÿ™ÿ≠ŸàŸäŸÑ ÿ•ŸÑŸâ ŸÖŸàÿ™ÿ± (Tensor) (ŸÖŸáŸÖ ÿ¨ÿØÿßŸã ŸÇÿ®ŸÑ ÿ•ÿØÿÆÿßŸÑŸáÿß ŸÑŸÑŸÜŸÖŸàÿ∞ÿ¨)
    transforms.ToTensor(),

    # 6. (ÿßÿÆÿ™Ÿäÿßÿ±Ÿä ŸÑŸÉŸÜ ŸÖŸàÿµŸâ ÿ®Ÿá) ÿßŸÑÿ™ÿ∑ÿ®Ÿäÿπ: Ÿäÿ≥ÿ™ÿÆÿØŸÖ ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÇŸäÿßÿ≥Ÿäÿ© ŸÑŸÄ ImageNet
    # transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```
from torchvision.datasets import ImageFolder
import os

base_dir = './images_dataSAT/'

# ÿ™ÿ≠ŸÖŸäŸÑ ŸÖÿ¨ŸÖŸàÿπÿ© ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ Ÿàÿ™ÿ∑ÿ®ŸäŸÇ ÿßŸÑÿ™ÿ≠ŸàŸäŸÑÿßÿ™
imagefolder_dataset = ImageFolder(
    root=base_dir,
    transform=custom_transform # ÿ™ÿ∑ÿ®ŸäŸÇ ÿÆÿ∑ ÿßŸÑÿ£ŸÜÿßÿ®Ÿäÿ® ÿßŸÑÿ∞Ÿä ÿ™ŸÖ ÿ•ŸÜÿ¥ÿßÿ§Ÿá ŸÅŸä ÿßŸÑŸÖŸáŸÖÿ© 1
)

print(f"‚úÖ ÿ™ŸÖ ÿ™ÿ≠ŸÖŸäŸÑ ŸÖÿ¨ŸÖŸàÿπÿ© ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™. ÿ•ÿ¨ŸÖÿßŸÑŸä ÿ≠ÿ¨ŸÖ ÿßŸÑŸÖÿ¨ŸÖŸàÿπÿ©: {len(imagefolder_dataset)}")
```
